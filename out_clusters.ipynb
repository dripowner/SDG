```python
import pandas as pd
```

```python
df = pd.read_csv("clustering.csv")
```

```python
df = df.set_index("ЦУР")
```

```python
df
```

```python
"\n".join(list(df[df["AgglomerativeClustering"] == 0].index))
```

```python
with open("agglomerative_clusters.txt", "w", encoding="utf-8") as f:
    for i in df["AgglomerativeClustering"].unique():
        f.write(f"\nКластер {i + 1}\n")
        f.write("\n".join(list(df[df["AgglomerativeClustering"] == 0].index)))
```

```python
with open("kmeans_clusters.txt", "w", encoding="utf-8") as f:
    for i in df["KMeans"].unique():
        f.write(f"\nКластер {i + 1}\n")
        f.write("\n".join(list(df[df["KMeans"] == 0].index)))
```
